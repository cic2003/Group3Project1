```python
#This is the file where I combine the Homeless_count.csv and the Clean_census_data.csv 
#Into one data file and I will save it into a new csv file as output 
#so now I will import some files such as pandas to read the csv file.
#I also import the numpy function to do aggregation and also matplotlib to create 
#charts and other data visualization types.
%matplotlib notebook
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Next I want to get the csv file I want to import by using the read_csv function/method 
#I am reading the homeless_veterans_data.csv and the clean_census_data.csv

homeless_veterans = pd.read_csv("Homeless_Count2018.csv", usecols=['State','Overall Homeless, 2018','Homeless Veterans, 2018','Sheltered ES Homeless Veterans, 2018','Sheltered TH Homeless Veterans, 2018','Sheltered SH Homeless Veterans, 2018','Sheltered Total Homeless Veterans, 2018','Unsheltered Homeless Veterans, 2018'], encoding="utf-8")
homeless_veterans.head(15)
```

```python
#This is another csv file I wanted to pull so I can work on the data
census_data = pd.read_csv("clean_census_data.csv", encoding="utf-8")
census_data.head(63)
```

```python
#This is the section where I drop the rows that provide no use to me
#This is especially true for PR since there was no data for homeless veterans

census_data = census_data.drop([0,1,2,3,4,56], axis = 0)
```

```python
#Then at this point I want to merge the data so I can aggregate data from one data frame.
#I might have to create an index key of sorts so they can have some commonality to be able to merge
#this is the section where I can see if index has a type 
type(homeless_veterans.index)
type(census_data.index)
```

```python
#Next I merge the data sets 
merge_veterans_data = pd.merge(census_data, homeless_veterans, how='inner', on='State')
merge_veterans_data.head(56)
```

```python
#Here I want to get a percentage of the homelessness in the USA . 
#So I need to get the total number of homeless in the USA and divide that by the total populatio
#of the USA.
#here I am setting the index as states so I can pull information by state and all that gobblebygook

new_data_frame = merge_veterans_data.set_index("State")
new_data_frame
```

```python
new_data_frame.dtypes
```

```python
#I had to rename the columns because python does not like spaces or commas in column names
#as far as I can tell.

homeless_df = new_data_frame.rename(columns={'NAME':'Abbreviation', 'POP_ESTIMATE_2018':'Pop_Est', 'Overall Homeless, 2018':'Tot_Homeless',
       'Homeless Veterans, 2018':'Homeless_Vet_Total', 'Sheltered ES Homeless Veterans, 2018':'ES_Homeless_Vet',
       'Sheltered TH Homeless Veterans, 2018':'TH_Homeless_Vet',
       'Sheltered SH Homeless Veterans, 2018':'SH_Homeless_Vet',
       'Sheltered Total Homeless Veterans, 2018':'Tot_Shelter_Homeless_Vet',
       'Unsheltered Homeless Veterans, 2018':'Unshelter_Homeless_Vet'})
homeless_df

```

```python
#This is the section where I want to see the data types of the columns and I will have an idea 
#if there is a data type needs to be converted.
homeless_df.dtypes
```

```python
#This is the point where I need to replace the comma in the columns 
#In order to convert the datatype
homeless_df['Tot_Homeless'] = homeless_df['Tot_Homeless'].str.replace(',','')
homeless_df['Homeless_Vet_Total'] = homeless_df['Homeless_Vet_Total'].str.replace(',','')
homeless_df['ES_Homeless_Vet'] = homeless_df['ES_Homeless_Vet'].str.replace(',','')
homeless_df['TH_Homeless_Vet'] = homeless_df['TH_Homeless_Vet'].str.replace(',','')
homeless_df['Tot_Shelter_Homeless_Vet'] = homeless_df['Tot_Shelter_Homeless_Vet'].str.replace(',','')
homeless_df['Unshelter_Homeless_Vet'] = homeless_df['Unshelter_Homeless_Vet'].str.replace(',','')
```

```python
#This is to check if the data type for my sake

homeless_df.dtypes
```

```python
#This is the stage where I want to convert the data type into integer for the rest of the columns 
homeless_df['Pop_Est'] = homeless_df['Pop_Est'].astype(float)
homeless_df['Tot_Homeless'] = homeless_df['Tot_Homeless'].astype(float)
homeless_df['Homeless_Vet_Total'] = homeless_df['Homeless_Vet_Total'].astype(float)
homeless_df['ES_Homeless_Vet'] = homeless_df['ES_Homeless_Vet'].astype(float)
homeless_df['TH_Homeless_Vet'] = homeless_df['TH_Homeless_Vet'].astype(float)
homeless_df['SH_Homeless_Vet'] = homeless_df['SH_Homeless_Vet'].astype(float)
homeless_df['Tot_Shelter_Homeless_Vet'] = homeless_df['Tot_Shelter_Homeless_Vet'].astype(float)
homeless_df['Unshelter_Homeless_Vet'] = homeless_df['Unshelter_Homeless_Vet'].astype(float)
```

```python
#Again checking the datatype
homeless_df.dtypes
```

```python
#then I will extract information based on the US population.  This should be fun

US_pop_2018 = homeless_df["Pop_Est"].sum()
US_pop_2018

Homeless_pop_2018 = homeless_df["Tot_Homeless"].sum()
Homeless_pop_2018

print(US_pop_2018)
print(Homeless_pop_2018)

US_pop = homeless_df["Pop_Est"]
Homeless_pop = homeless_df["Tot_Homeless"]
```

```python
#This is the percentage of the population that are homeless

percentage = (Homeless_pop_2018/US_pop_2018 * 100).round(2)
percentage
```

```python
#Now I want to calculate the percentages of homeless population out of the general population
colors = ["yellowgreen", "red"]
labels = ["US Estimated Population in 2018", "Total Homeless Population in 2018"]
size = [US_pop_2018, Homeless_pop_2018]
#plt.title("The Percantage of the Total Homelessness in the US, 2018")
explode = (0.1, 0.4)

fig, ax = plt.subplots()
ax.pie(size,
       explode = explode,
       labels = labels,
       autopct = '%1.1f%%',
       shadow = False,
       startangle = 45)
ax.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
ax.set_title('The Percentage of the Total Homelessness in the US, 2018')
```
