```python
#This is the file where I combine the Homeless_count.csv and the Clean_census_data.csv 
#Into one data file and I will save it into a new csv file as output 
#so now I will import some files such as pandas to read the csv file.
#I also import the numpy function to do aggregation and also matplotlib to create 
#charts and other data visualization types.
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib
```

```python
#Next I want to get the csv file I want to import by using the read_csv function/method 
#I am reading the homeless_veterans_data.csv and the clean_census_data.csv

homeless_veterans = pd.read_csv("Homeless_Count2018.csv", 
                                usecols=['State','Homeless Veterans, 2018',
                                         'Sheltered Total Homeless Veterans, 2018',
                                         'Unsheltered Homeless Veterans, 2018'], 
                                encoding="utf-8")
homeless_veterans.head(15)
```

```python
#I want to drop the row containing the US row because I want to calculate that data on later
homeless_veterans = homeless_veterans.drop([0])
homeless_veterans
```

```python
homeless_veterans.dtypes
```

```python
#Reading the csv file for bed count and creating the dataframe to use in the project

bed_count = pd.read_csv("HIC_bedcount.csv", 
                        usecols=['State', 'Dedicated Veteran Beds (ES, TH, SH)'], 
                        encoding="utf-8")
bed_count.head(63)
```

```python
#Here I am dropping the row for the total because it will mess up my data 
bed_count = bed_count.drop([51])
bed_count
```

```python
bed_count.dtypes
```

```python
#Here I am merging the two tables together to make one dataframe to 
merge_vets_beds = pd.merge(homeless_veterans,bed_count, how='left', on='State')
merge_vets_beds
```

```python
##merge_vets_beds = merge_vets_beds.set_index("State")
##merge_vets_beds.index
```

```python
homeless_df = merge_vets_beds
homeless_df.head()
```

```python
homeless_df.index
```

```python
print (homeless_df.dtypes)
```

```python
new_homeless = homeless_df.rename(columns={'State':'State',
                                           'Homeless Veterans, 2018':'Homeless_Vets_Tot',
                                           'Sheltered Total Homeless Veterans, 2018':'Sheltered_Tot',
                                           'Unsheltered Homeless Veterans, 2018':'Unsheltered_Tot',
                                           'Dedicated Veteran Beds (ES, TH, SH)':'Total_Beds'})
new_homeless.head()
```

```python
new_homeless.dtypes
```

```python
new_homeless['State'] = new_homeless['State'].str.replace(',','')
new_homeless['Homeless_Vets_Tot'] = new_homeless['Homeless_Vets_Tot'].str.replace(',','')
new_homeless['Sheltered_Tot'] = new_homeless['Sheltered_Tot'].str.replace(',','')
new_homeless['Unsheltered_Tot'] = new_homeless['Unsheltered_Tot'].str.replace(',','')
```

```python
new_homeless.dtypes
```

```python
new_homeless['State'] = new_homeless['State'].astype(str)
new_homeless['Homeless_Vets_Tot'] = new_homeless['Homeless_Vets_Tot'].astype(float)
new_homeless['Sheltered_Tot'] = new_homeless['Sheltered_Tot'].astype(float)
new_homeless['Unsheltered_Tot'] = new_homeless['Unsheltered_Tot'].astype(float)
new_homeless['Total_Beds'] = new_homeless['Total_Beds'].astype(float)
```

```python
new_homeless.dtypes
```

```python
# Setting the positions and width for the bars
pos = list(range(len(new_homeless['Sheltered_Tot']))) 
width = 0.25 
    
# Plotting the bars
fig, ax = plt.subplots(figsize=(15,5))

# Create a bar with pre_score data,
# in position pos,
plt.bar(pos, 
        #using df['pre_score'] data,
        new_homeless['Sheltered_Tot'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#EE3224', 
        # with label the first value in first_name
        label=new_homeless['State'][0]) 

# Create a bar with mid_score data,
# in position pos + some width buffer,
plt.bar([p + width for p in pos], 
        #using df['mid_score'] data,
        new_homeless['Unsheltered_Tot'],
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#F78F1E', 
        # with label the second value in first_name
        label=new_homeless['State'][1]) 

# Create a bar with post_score data,
# in position pos + some width buffer,
plt.bar([p + width*2 for p in pos], 
        #using df['post_score'] data,
        new_homeless['Total_Beds'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#FFC222', 
        # with label the third value in first_name
        label=new_homeless['State'][2]) 

# Set the chart's title
ax.set_title('The Number of Sheltered and Unsheltered Homeless in Relations to Number of Dedicated Beds in Each State')

# Set the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(new_homeless['State'])

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, max(new_homeless['Sheltered_Tot'] + new_homeless['Unsheltered_Tot']+ new_homeless['Total_Beds'])] )

# Adding the legend and showing the plot
plt.legend(['Sheltered Homeless Veterans', 'Unsheltered Homeless Veterans', 'Dedicated Beds for Veterans'], loc='upper left')
plt.grid()
plt.show()
```
